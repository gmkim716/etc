# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고', '능남고':'무슨고'})
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace = True)
df
```

```python
df
```

```python
df['SW특기'].str.lower()
```

```python
df['특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail' # 결과 column을 추가하고 전체 데이터는 Fail로 초기화
df
```

```python
df.loc[df['총합'] > 400, '결과'] = 'Pass' # 총합이 400보다 큰 데이터에 대해서 결과를 Pass로 업데이트
df
```

## Column 삭제

```python
df.drop(columns=['총합']) # 총합 column을 삭제
```

```python
df.drop(columns=['국어','영어','수학']) # 국어, 영어, 수학 column을 삭제
```

```python
df
```

## Row 삭제

```python
df.drop(index='4번') # 4번 학생 데이터 row를 삭제
```

```python
filt = df['수학'] < 80 # 수학 점수 80점 미만 학생 필터링
df
```

```python
df[filt].index
```

```python
df.drop(index=df[filt].index)
df
```

```python
df.drop(columns=['특기'], inplace=True)
df
```

## Row 추가

```python
df.loc['9번'] = ['이정환','해남고등학교',184,90,90,90,90,90,'kotlin',450,'Pass'] # 새로운 row 추가
df
```

## Cell 수정

```python
df.loc['4번','SW특기'] = 'Python' # 4번 학생의 SW특기 데이터를 Python으로 변경
df
```

```python
df.loc['5번',['학교', 'SW특기']] = ['능남고등학교', 'C'] # 5번 학생의 학교는 능남고등학교로, SW특기는 C로 변경
df
```

## Column 순서 변경

```python
cols = list(df.columns)
cols
```

```python
df = df[[cols[-1]] + cols[0:-1]] # 맨 뒤에 있는 결과 column을 앞으로 가져오고, 나머지 column들과 합쳐서 순서 변경
df
```
